## Dropping columns that are created in a redundant manner

```python
ss=ss.drop(["Unnamed: 0","Unnamed: 0.1"],axis=1)
```

## Modifying the dataset in order to carry out "Group By" operation

```python
ss["Date"]=ss["Date"].apply(lambda x:2012 if "2012" in x else 2013)
```

## Creation of 2012 month dataset

```python
df2012=pd.read_csv("stocksdataviz.csv")
```

```python
def check(x):
    if x.startswith("2012"):
        return x


df2012["Date"]=df2012["Date"].apply(check)
```

```python
df2012
```

```python
df2012=df2012.dropna(axis=0)
```

```python
df2012
```

```python
def checktwice(x):
    if "-01-" in x:
        return "January"
    elif "-02-" in x:
        return "February"
    elif "-03-" in x:
        return "March"
    elif "-04-" in x:
        return "April"
    elif "-05-" in x:
        return "May"
    elif "-06-" in x:
        return "June"
    elif "-07-" in x:
        return "July"
    elif "-08-" in x:
        return "August"
    elif "-09-" in x:
        return "September"
    elif "-10-" in x:
        return "October"
    elif "-11-" in x:
        return "November"
    elif "-12-" in x:
        return "December"
```

```python
df2012["Date"]=df2012["Date"].apply(checktwice)
```

```python
df2012
```

```python
dfmonth=df2012.groupby("Date").sum()
```

```python
dfmonth.drop(["Unnamed: 0","Unnamed: 0.1"],axis=1,inplace=True)
```

```python
dfmonth
```

```python
dfmonth.to_csv("2012months.csv")
```
